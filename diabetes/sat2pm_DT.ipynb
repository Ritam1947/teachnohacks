```python
col_names=['pregnant','glucose','bp','skin','insulin','bmi','pedigree','age','label']
```

```python
import pandas as pd
df=pd.read_csv('pima-indians-diabetes.csv',names=col_names)
df
```

```python
feature_col=['pregnant','glucose','bp','skin','insulin','bmi','pedigree','age']
```

```python
x=df[feature_col]
x
```

```python
y=df.label
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
cl=DecisionTreeClassifier(criterion="gini",max_depth=3)
cl.fit(xtrain,ytrain)
```

```python
ypred=cl.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(ytest,ypred)
```

```python
accuracy_score(ytest,ypred)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(cl)
```

```python
cl1=DecisionTreeClassifier(criterion='entropy',max_depth=3)
cl1.fit(xtrain,ytrain)
```

```python
ypred1=cl1.predict(xtest)
```

```python
confusion_matrix(ytest,ypred1)
```

```python
accuracy_score(ytest,ypred1)
```

```python
tree.plot_tree(cl1)
```
